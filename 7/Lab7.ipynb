***Модель полиномиальной регрессии (polynomial regression)*** - это обобщенная линейная модель с параметризованной степенью многочлена.

![img](https://russianblogs.com/images/753/361716febc5a15ae65cdba4394db1f69.png)

Многочлен 1-й степени:

$$ y=ax+b $$ 

Многочлен 3-й степени:
$$ y = ax^3 + bx^2 + cx + d $$

Несмотря на то, что мы можем использовать полиномиальную регрессию для моделирования нелинейных связей, она по-прежнему рассматривается как модель множественной линейной регрессии, ввиду 
линейных коэффициентов регрессии.

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures  # Класс преобразователь
```

[Ссылка на документацию](https://scikit-learn.org/stable/modules/preprocessing.html#polynomial-features)

```python
# Моделируем данные
X = np.array([258.0, 270.0, 294.0, 320.0, 342.0, 368.0, 396.0, 446.0, 480.0, 586.0])[:, np.newaxis] # тоже самое, что reshape -1;1
y = np.array([236.4, 234.4, 252.8, 298.6, 314.2, 342.2, 360.8, 368.0, 391.2, 390.8])
```

```python
plt.scatter(X, y) 
```

```python
lr = LinearRegression() # Линейная регрессия
lr.fit(X, y) # Обучим линейную регрессию
```

```python
X_ = np.arange(250, 600, 10)[:, np.newaxis] # Точки для предсказания
y_lr = lr.predict(X_) # Предсказываем y для линейной регрессии
```

```python
X_
```

```python
plt.scatter(X, y) 
plt.plot(X_, y_lr) 
```

```python
pr = LinearRegression() # Полиномиальная регрессия
```

```python
quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X)       # Преобразуем данные
```

```python
X_quad
```

```python
pr.fit(X_quad, y) # Обучаем полиномиальную регрессию
```

```python
y_pr = pr.predict(quadratic.fit_transform(X_)) # Предсказываем y для полиномиальной регрессии
```

```python
plt.scatter(X, y, label = 'тренировочные точки') 
plt.plot(X_, y_lr, label = 'линейная подгонка') 
plt.plot(X_, y_pr, label = 'квадратичная подгонка') 
plt.legend(loc='upper left') 
plt.show()
```

Задание 1. Оцените полученные модели с помощью коэффициента детерминации и MSE. Что можно сказать о качестве моделей?

Оценка R2. Суть его работы заключается в измерении количества отклонений в прогнозах, объясненных набором данных. Это разница между выборками в наборе данных и прогнозами, сделанными моделью.
Если значение оценки R в квадрате равно 1, то модель считается идеальной. Если ее значение равно 0, то модель будет плохо работать с неизвестным набором данных. То есть, чем ближе значение оценки R в квадрате к 1, тем точнее обучается модель.


```python
print("Линейная = ", sklearn.metrics.r2_score(y, y_lr[0:len(y):]))
print("Квадратичная = ", sklearn.metrics.r2_score(y, y_pr[0:len(y):]))
```

MSE - Среднее квадратичное отклонение.
Измеряет среднее квадратическое различие между оценочными значениями и истинными значениями.
Это функция риска, соответствующая ожидаемому значению квадрата потери ошибок. Это всегда неотрицательно, и значения, близкие к нулю, лучше. 
MSE является вторым моментом ошибки (относительно источника) и, таким образом, включает в себя как дисперсию оценки, так и ее смещение.

```python
print("Линейная MSE = ", sklearn.metrics.mean_squared_error(y, y_lr[0:len(y):]))
print("Квадратичная MSE = ", sklearn.metrics.mean_squared_error(y, y_pr[0:len(y):]))
```

Задание 2. Постройте полиномиальную модель, описывающую следующую зависимость:

```python
X = np.array([10, 12, 15, 20, 25, 30, 34, 40, 47, 54, 57])[:, np.newaxis]
y = np.array([80, 75, 70, 63, 65, 70, 76, 85, 90, 92, 87 ])


plt.scatter(X, y) # Изображаем точки на графике
plt.show()

```

```python
lr = LinearRegression()
lr.fit(X, y)
X_n = np.arange(10, 70, 10)[:, np.newaxis] # Точки для предсказания
y_lr_n = lr.predict(X_n)



pr  = LinearRegression()
quadratic = PolynomialFeatures(degree=5)
X_quad = quadratic.fit_transform(X)
pr.fit(X_quad, y)
y_pr = pr.predict(quadratic.fit_transform(X_n))
plt.scatter(X, y, label = "тренировочные точки")
plt.plot(X_n, y_lr_n, label = "линейная подгонка", color = "green")
plt.plot(X_n, y_pr, label = "квадратичная подгонка", color = "red")
plt.legend(loc = "upper left")
plt.show()

```

Задание 3. Какую степень полинома вы использовали? Попробуйте изменить степень полинома, например, на 2, 5, 7, 15? Как меняется результат?

Использовалась 5-я степень полинома, т.к. она более точно описывает связь между x и y;
Если менять степень, то меняется описание данных нашей моделью(появляются более серьезные ошибки предсказания). 
Ниже представлены варианты степени от 20 до 2

```python
deg = 2
while deg < 20:
    print("Degree is ", deg)
    pr  = LinearRegression()
    quadratic = PolynomialFeatures(degree=deg)
    X_quad = quadratic.fit_transform(X)
    pr.fit(X_quad, y)
    y_pr = pr.predict(quadratic.fit_transform(X_n))
    plt.scatter(X, y, label = "тренировочные точки")
    plt.plot(X_n, y_lr_n, label = "линейная подгонка", color = "green")
    plt.plot(X_n, y_pr, label = "квадратичная подгонка", color = "red")
    plt.legend(loc = "upper left")
    plt.show()
    deg +=1
```
