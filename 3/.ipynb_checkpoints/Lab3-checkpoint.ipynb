1. Загрузите данные из файла в объект DataFrame, Добавьте заголовки к
столбцам: «index», «year», «month», «day», «min_t», «average_t», «max_t»,
«rainfall».
Расшифровка:
• index – индекс ВМО,
• year – год,
• month – месяц,
• day – день,
• min_t – минимальная температура воздуха,
• average_t – средняя температура воздуха,
• max_t – максимальная температура воздуха,
• rainfall – количество осадков.

```python
import pandas

data = pandas.read_table('data.txt', sep=';', names = ['index', 'year', 'month', 'day', 'min_t', 'average_t', 'max_t', 'rainfall'])

data['min_t'] = data['min_t'].str.replace(' ', '')
data['average_t'] = data['average_t'].str.replace(' ', '')
data['max_t'] = data['max_t'].str.replace(' ', '')
data['rainfall'] = data['rainfall'].str.replace(' ', '')

data = data.apply(pandas.to_numeric)

data.head()
```

```python
# 2. Удалите столбец index
del data['index']
data.head()
```

3. Используя метод info(), ответьте на вопросы:

```python
# 3.1. Есть ли в данных пропущенные значения?
print(data.head())
print(data.info())
# Ответ: Есть

# 3.2. В каком столбце данных больше всего пропущенных значений?
data.isnull().sum()
# Ответ: max_t
```

4. В данных за какой год больше всего пропусков?

```python
result = data.groupby('year', as_index = False)[['month', 'day', 'min_t', 'average_t', 'max_t', 'rainfall']].count()
str = result[result['max_t']==result['max_t'].min()]
print(str)
```

5. Объедините столбцы «year», «month» и «day» в один столбец «date» в формате гггг-мм-дд (2000-01-20). Данные в новом столбце должны иметь формат datetime.

 6. Для каждого наблюдения рассчитайте размах температур (разность максимальной и минимальной суточных температур) и количество предшествующих ему дней без осадков 
(используйте циклы Python и условный оператор).


```python

print(data)

count = 0
countRows = 10
for row in data.itertuples():
    if (row[8]) == 0.0:
        count += 1
    else: 
        count = 0
        
    print("Индекс:", int(row[0]), "\nРазмах температур:", (float(row[7]) - float(row[5])), "\nКоличество дней без осадков:", count, "\n") 
    if int(row[0]) + 1 > countRows:
        break
```

7. Определите самый длинный период засухи.

```python
count = 0
maxDayIndex = 0
maxDayRain = 0

for row in data.itertuples():
    if float(row[8]) == 0.0:
        count += 1
    else: 
        count = 0

    if (count > maxDayRain):
        maxDayRain = count

print("Количество дней засухи: ", maxDayRain) 
```

```python
# 8. Для каждого года вычислите среднегодовую температуру и общее количество осадков. Запишите результаты в объекты Series.

agg_func_math = {
    'average_t': ['mean'],
    'rainfall':['sum']
}
result = data.groupby(['year']).agg(agg_func_math).round(2)
print(result)

# 8.1. Какой год можно считать самым теплым? Какой самым холодным?

print(result['average_t'].max())
print(result['average_t'].min())
print('\n')

# 8.2. В какой год выпало больше всего осадков? В какой меньше всего?
# Используя запись имя_серии.plot() вы можете построить график и посмотреть как изменялась температура. С помощью имя_серии.plot.bar() можно отобразить на столбиковой диаграмме количество осадков,
# выпавших в каждый год.

print('\n')
print(result['rainfall'].max())
print(result['rainfall'].min())
```
