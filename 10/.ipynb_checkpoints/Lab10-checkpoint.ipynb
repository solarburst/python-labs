Цель: обнаружить с помощью DBSCAN в наборе фотографий
«нетипичные» фото.
1) С помощью метода главных компонент извлеките из X_people 100
главных компонент (PCA(n_components=100, whiten=True, random_state=0)).
2) Осуществите кластеризацию методом DBSCAN, используя разные
значения параметра eps (1, 3, 5, 7, 9, 11, 13, 15, 20, 30), min_samples=3, и
проанализируйте какие структуры алгоритму удастся выявить в этом наборе
данных. Какие изображения попали в «шум»?

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
image_shape = people.images[0].shape
# Отобразим несколько изображений
fig, axes = plt.subplots(2, 5, figsize=(12, 5), subplot_kw={'xticks':
(), 'yticks': ()})
for target, image, ax in zip(people.target, people.images,
axes.ravel()):
 ax.imshow(image, cmap = 'gray')
 ax.set_title(people.target_names[target])
# Для анализа будем использовать не более 50 изображений каждого

mask = np.zeros(people.target.shape, dtype=bool)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1
X_people = people.data[mask]
y_people = people.target[mask]
# Для получения большей стабильности масштабируем шкалу оттенков
#серого так, чтобы значения были в диапазоне от 0 до 1 вместо
#использования шкалы значений от 0 до 255
X_people = X_people / 255.



X_train, X_test, y_train, y_test = train_test_split(
 X_people, y_people, stratify=y_people, random_state=0)
# строим KNeighborsClassifier с одним соседом
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Правильность на тестовом наборе для 1-nn: {:.2f}".format(knn.score(X_test, y_test)))
pca = PCA(n_components=100, whiten=True, random_state=0).fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
print("Обучающие данные после PCA: {}".format(X_train_pca.shape))
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_pca, y_train)
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test_pca, y_test)))
print("форма pca.components_: {}".format(pca.components_.shape))
fix, axes = plt.subplots(3, 5, figsize=(15, 12),
subplot_kw={'xticks': (), 'yticks': ()})
for i, (component, ax) in enumerate(zip(pca.components_, axes.ravel())):
 ax.imshow(component.reshape(image_shape),
 cmap='viridis')
 ax.set_title("{}. component".format((i + 1)))
    



```

```python
X_pca = pca.transform(X_people)    

listEps = (1, 3, 5, 7, 9, 11, 13, 15, 20, 30);   

for i in listEps:
    print("EPS is ", i)
    dbscan = DBSCAN(min_samples=3, eps=i)
    labels = dbscan.fit_predict(X_pca)
    print("Уникальные метки: {}".format(np.unique(labels)))
    noise = X_people[labels==-1]
    fig, axes = plt.subplots(3, 9, subplot_kw={'xticks': (), 'yticks': ()},
    figsize=(12, 4))
    for image, ax in zip(noise, axes.ravel()):
        ax.imshow(image.reshape(image_shape), vmin=0, vmax=1)
```
